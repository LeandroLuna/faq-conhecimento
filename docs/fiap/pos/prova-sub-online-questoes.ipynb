# **Lendo o arquivo**

```python
import pandas as pd

df = pd.read_excel("base.xlsx")

df.head()
```

# **Verificando e removendo as duplicadas**

```python
# linhas_duplicadas = df[df.duplicated()]
# print(linhas_duplicadas)
df_questoes = df.drop(['CURSO', 'Insira seu nome', 'Insira seu e-mail'], axis=1)
df_questoes = df_questoes.drop_duplicates()
df_questoes.head()
```

# **Quebrando as questões e alternativas por linha**

```python


num_linhas = df_questoes.shape[0]
num_colunas = df_questoes.shape[1]

questoes = []
questao = []
i = 0
j = 1
cont = 0

for i in range(0, num_linhas):
  for j in range(1, num_colunas):

    if(cont == 0):
      questao = []
      questao.append(df_questoes.iloc[i, 0])

    questao.append(df_questoes.iloc[i, j])
    j+=1
    cont+=1

    if(cont == 6):
      cont = 0
      questoes.append(questao)

  j = 1
  i+=1

print(len(questoes))
print(questoes[0][0])

```

# **Gerando o script sql**

```python
script = ("DECLARE @insertResultado TABLE (Codigo INT); \n"
"DECLARE @codDisciplina INT; \n"
"DECLARE @codQuestaoGerada INT; \n"
"DECLARE @codCategoria INT = (SELECT Codigo FROM site_fiap..POQuestaoCategoria WHERE Descricao = 'Revisão de Notas e Faltas Online - MBA'); \n\n\n");

for i in range(0, len(questoes)):

  index = str(i+1)
  disciplina = str(questoes[i][0]).replace("'", "''")
  enunciado = str(questoes[i][1]).replace("'", "''")
  alternativa1 = str(questoes[i][2]).replace("'", "''")
  alternativa2 = str(questoes[i][3]).replace("'", "''")
  alternativa3 = str(questoes[i][4]).replace("'", "''")
  alternativa4 = str(questoes[i][5]).replace("'", "''")
  alternativa5 = str(questoes[i][6]).replace("'", "''")

  novaQuestao = ("--QUESTAO "+index+"---------------------------------------------------------------------------------------------------------------------\n\n"

  "SELECT TOP 1 @codDisciplina = Disciplina_2004.Codigo FROM site_fiap..Disciplina_2004 WHERE Descricao = '"+disciplina+"' ORDER BY Codigo DESC; \n\n"

  "INSERT INTO site_fiap..POQuestao (CodigoDisciplina, Serie, Ano, Pergunta, Ativo, Anulada) \n"
  "OUTPUT inserted.Codigo INTO @insertResultado \n"
  "VALUES (@codDisciplina, 1, YEAR(GETDATE()), '"+enunciado+"', 1, 0); \n\n"

  "SET @codQuestaoGerada = (SELECT TOP 1 Codigo FROM @insertResultado ORDER BY Codigo DESC); \n\n"

  "INSERT INTO site_fiap..POQuestaoCategoriaQuestoes VALUES (@codCategoria, @codQuestaoGerada); \n\n"

  "INSERT INTO site_fiap..POQuestaoAlternativa (CodigoQuestao, Alternativa, Correta, Ativo) \n"
  "VALUES (@codQuestaoGerada, '"+alternativa1+"', 1, 1); \n\n"

  "INSERT INTO site_fiap..POQuestaoAlternativa (CodigoQuestao, Alternativa, Correta, Ativo) \n"
  "VALUES (@codQuestaoGerada, '"+alternativa2+"', 0, 1); \n\n"

  "INSERT INTO site_fiap..POQuestaoAlternativa (CodigoQuestao, Alternativa, Correta, Ativo) \n"
  "VALUES (@codQuestaoGerada, '"+alternativa3+"', 0, 1); \n\n"

  "INSERT INTO site_fiap..POQuestaoAlternativa (CodigoQuestao, Alternativa, Correta, Ativo) \n"
  "VALUES (@codQuestaoGerada, '"+alternativa4+"', 0, 1); \n\n"

  "INSERT INTO site_fiap..POQuestaoAlternativa (CodigoQuestao, Alternativa, Correta, Ativo) \n"
  "VALUES (@codQuestaoGerada, '"+alternativa5+"', 0, 1); \n\n\n\n" )

  script = script + novaQuestao
  i += 1

with open("INSERT_QUESTOES.sql", "w") as arquivo:
    arquivo.write(script)

print("Arquivo SQL salvo com sucesso!")
```
